## Sentiment of 10 Best Foods for You

```python
# Storing required values as dataframe in new variable
NewApp = adf[adf['App'] == '10 Best Foods for You']
NewApp
```

```python
# Grouping_by_sentiments_to_find_total_positive_negative_and_neutral_sentiments
topAppSentiments = pd.DataFrame(NewApp['Sentiment'].value_counts()).reset_index()
topAppSentiments
```

```python
# Finding_the_percentage
topAppSentiments['Percent'] = (topAppSentiments['Sentiment'] / topAppSentiments['Sentiment'].sum()) *100
topAppSentiments
```

```python
plt.pie(topAppSentiments['Sentiment'], labels = topAppSentiments['index'], autopct = '%1.1f%%', colors = ['pink', 'blue', 'green'], textprops={'fontsize': 14}, explode = [0.05, 0, 0])
plt.show()
```

### If we analysis the above pie chart then we conclude that there is high percentage (83.5%) of Positive sentiment for 10 Best Foods for You and for Negative there is 5.2% and for Neutral is 11.3%
